```python
#years list from 2011 to 2017, specify the period of research
years=[2011,2012,2013,2014,2015,2016,2017]
#players last name and dictionary, for looping purpose and record
players=["kohli","sharma","dhawan","root","amla"]
playerdict={"kohli":0,"sharma":1,"dhawan":2,"root":3,"amla":4}
#runs scored by players from 2011 to 2017 , year vise
kohlirun=[1381,1026,1268,1045,623,739,1460]
rohitrun=[611,168,1196,578,815,564,1293]
dhawanrun=[960,287,745,815,1162,69,51]
dhawanrun=[51,69,1162,815,745,287,960]
rootrun = [0,0,626,872,723,796,983]
amlarun=[862,511,1062,892,838,678,632]
amlarun=[632,678,838,892,1062,511,862]
#iinigs played by players from 2011 to 2017, year vise
kohliInn=[34,17,30,20,20,10,26]
rohitInn=[16,13,27,12,17,10,21]
dhawanInn=[4,4,25,18,20,5,22]
rootInn=[0,0,19,22,18,14,18]
amlaInn=[18,14,23,18,22,9,15]
amlaInn=[15,9,22,18,23,14,18]
#matrix creation of runs of all players, each row belongs to one player, and each column defines single year from 2011 to 2017.
import numpy as np
runs=np.array([kohlirun,rohitrun,dhawanrun,rootrun,amlarun])
#matrix creation of innings,each row belongs to one player and column defines single year from 2011 to 2017.
innings=np.array([kohliInn,rohitInn,dhawanInn,rootInn,amlaInn])
#To find the average runs scored by all players in specific years we divide runs by innings.
import warnings
warnings.filterwarnings=("ignore")
avg_runs=np.matrix.round(runs/innings)

```

```python
#Visualization of total runs scored in that period by the 5 players
from matplotlib import pyplot, pylab
import matplotlib.pyplot as plt
%matplotlib inline 
#defing a function customplot, which is very helpfull reduces the code, iterates over the give array-data and gives us visuals 
def customplot(data,player=players):
    dictcolor={"kohli":"magenta","sharma":"red","dhawan":"black","root":"red","amla":"orange"}
    dictls={"kohli":"--","sharma":"-","dhawan":"-.","root":"dotted","amla":"-"}
    dictmarkers={"kohli":"*","sharma":"o","dhawan":"v","root":",","amla":"+"}
    plt.rcParams['figure.figsize']=10,8
    plt.xlabel("YEARS")
    plt.ylabel("RUNS")
    
    for name in player:
        plot1=plt.plot(runs[playerdict[name]],c=dictcolor[name],ls=dictls[name],marker=dictmarkers[name],label=name,ms=8)
        plot1=plt.xticks(list(range(0,8)),years,rotation="vertical")
        plt.legend(loc='upper right')
        plt.savefig("runs")

customplot(runs)
```

```python
#Visualization of average runs scored in that period by the 5 players
from matplotlib import pyplot, pylab
import matplotlib.pyplot as plt
%matplotlib inline 
#defing a function customplot, which is very helpfull reduces the code, iterates over the give array-data and gives us visuals 
def customplot(data,player=players):
    dictcolor={"kohli":"magenta","sharma":"red","dhawan":"black","root":"red","amla":"orange"}
    dictls={"kohli":"--","sharma":"-","dhawan":"-.","root":"dotted","amla":"-"}
    dictmarkers={"kohli":"*","sharma":"o","dhawan":"v","root":",","amla":"+"}
    plt.rcParams['figure.figsize']=10,8
    plt.xlabel("YEARS")
    plt.ylabel("AVG. RUNS")
    
    for name in player:
        plot1=plt.plot(avg_runs[playerdict[name]],c=dictcolor[name],ls=dictls[name],marker=dictmarkers[name],label=name,ms=8)
        plot1=plt.xticks(list(range(0,8)),years,rotation="vertical")
        plt.legend(loc='upper right')
        plt.savefig("avgruns")

customplot(avg_runs)
```
